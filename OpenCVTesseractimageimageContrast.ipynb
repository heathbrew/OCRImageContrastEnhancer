```python
import cv2
import pytesseract
from pdf2image import convert_from_path
```

```python
pdf_file = 'table.pdf'
pages = convert_from_path(pdf_file, 350)

i = 1
for page in pages:
    image_name = "Page_" + str(i) + ".jpg"  
    page.save(image_name, "JPEG")
    i = i+1       
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Load the original image
image = cv2.imread('3.jpg')
```

```python
# Adjust contrast
contrast = 0.95  # Adjust the value as desired (between 0 and 1)
adjusted_image = np.clip(image.astype(np.float32) * contrast, 0, 255).astype(np.uint8)
```

```python
# Display the adjusted image
cv2_imshow( adjusted_image)
cv2_imshow( image)
cv2.waitKey(0)
cv2.destroyAllWindows()
# Save the adjusted image
cv2.imwrite('adjusted_image.jpg', adjusted_image)

# Display a message after saving
print("Adjusted image saved successfully.")
```
